```python
# Modulador FSK coerente

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

# Parmetros
fc = 2000  # Frequncia central (Hz)
fd = 1000  # Desvio de frequncia (Hz)
Tb = 0.001  # Durao de um bit (segundos)
num_bits = 10  # Nmero de bits
samples_per_bit = 200  # Amostras por bit
total_samples = samples_per_bit * num_bits  # Total de amostras

# Vetor de tempo total
t = np.linspace(0, Tb*num_bits, total_samples)

# Sequncia de bits aleatrios
data = (np.random.rand(num_bits) > 0.5)
data_sequence = np.repeat(data, samples_per_bit)

# Sinal FSK

# Na modulao BFSK coerente, a fase inicial das portadoras em t=0  a mesma.
phi = np.random.uniform(0, 2*np.pi)
c1_coerente = np.sin(2 * np.pi * (fc - fd) * t + phi)
c2_coerente = np.sin(2 * np.pi * (fc + fd) * t + phi)

# Se o bit for 1: uma portadora com frequncia f1 = (fc - fd)  transmitida.
# Se o bit for 0: uma portadora com frequncia f2 = (fc + fd)  transmitida.

fsk_coerente = data_sequence*c1_coerente + (1 - data_sequence)*c2_coerente


plt.figure(figsize=(10, 6))
plt.plot(t, data_sequence, linestyle='--', color='r', linewidth=4, label='Digital Signal')
plt.plot(t, fsk_coerente, label='FSK Signal')
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (volt)')
plt.title('FSK Modulation')
plt.legend()
plt.show()
```

```python
# Modulador FSK no coerente

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(12)

# Parmetros
fc = 2000  # Frequncia central (Hz)
fd = 1000  # Desvio de frequncia (Hz)
Tb = 0.001  # Durao de um bit (segundos)
num_bits = 10  # Nmero de bits
samples_per_bit = 200  # Amostras por bit
total_samples = samples_per_bit * num_bits  # Total de amostras

# Vetor de tempo total
t = np.linspace(0, Tb*num_bits, total_samples)

# Sequncia de bits aleatrios
data = (np.random.rand(num_bits) > 0.5)
data_sequence = np.repeat(data, samples_per_bit)

# Sinal FSK

# Na modulao BFSK no coerente, a fase inicial das portadoras em t=0  diferente.
phi1 = np.random.uniform(0, 2*np.pi)
phi2 = np.random.uniform(0, 2*np.pi)
c1_ncoerente = np.sin(2 * np.pi * (fc - fd) * t + phi1)
c2_ncoerente = np.sin(2 * np.pi * (fc + fd) * t + phi2)

# Se o bit for 1: uma portadora com frequncia f1 = (fc - fd)  transmitida.
# Se o bit for 0: uma portadora com frequncia f2 = (fc + fd)  transmitida.

fsk_ncoerente = data_sequence*c1_ncoerente + (1 - data_sequence)*c2_ncoerente


plt.figure(figsize=(10, 6))
plt.plot(t, data_sequence, linestyle='--', color='r', linewidth=4, label='Digital Signal')
plt.plot(t, fsk_ncoerente, label='FSK Signal')
plt.grid(True)
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude (volt)')
plt.title('FSK Modulation')
plt.legend()
plt.show()
```
